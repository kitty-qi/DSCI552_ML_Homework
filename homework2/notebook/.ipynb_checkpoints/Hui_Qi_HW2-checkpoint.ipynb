### (c) Simple Linear Regression

### AT predictor

```python
# https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.OLS.html
# https://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.OLSResults.outlier_test.html
univariate_coeff = []

i= 0 
X = df.iloc[:,i] # AT
Y = df.iloc[:,-1]
X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
print(f"For the simple linear regression model between {df.columns[i]} and {df.columns[-1]},\
the beta0(intercept) is {results.params[0]} while beta1(slope) is {results.params[1]}.\n",
f"It implies that an 1 unit increase in x will decrease y by {abs(results.params[1])} units when other variables are fixed.")
print("\n")
print(f"The p-vlue for {df.columns[i]} is almost 0.000, which is less than significant level of 0.05. \
Therefore, we could reject the null hypothesis and conclude that AT is statistically significant in predicting PE in this model.")
print("\n")
print("R-squared is 0.899. There is a strong linear relationship between AT and PE.")
print(results.summary())


```

```python
# https://stackoverflow.com/questions/42261976/how-to-plot-statsmodels-linear-regression-ols-cleanly
p = results.params
univariate_coeff.append(p[1])
sig_level = 0.05
# https://stackoverflow.com/questions/42261976/how-to-plot-statsmodels-linear-regression-ols-cleanly
# https://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.OLSResults.outlier_test.html
stud_res = results.outlier_test()

x= df.AT
y = df.PE
plt.scatter(x,y)
plt.plot(x, p[1]*x+p[0], color='red')
plt.title("Simple Linear Regression for PE vs AT")
plt.show()
```

### (f) Nonlinear Association

```python
def Polyregression(X,y):
    X = X.array.reshape(-1, 1)
    poly = PolynomialFeatures(degree = 3)
    X_poly = poly.fit_transform(X)
  
    poly_reg_model = sm.OLS(y,X_poly)
    poly_result = poly_reg_model.fit()
    return poly_result.summary()
```

```python
print("All beta has p-value almost 0, which is less than 0.05. \
Therefore, we could rejetct the null hypothesis and conclude that \
there is evidence of nonlinear association between the predictor AT and the response PE. \n")
Polyregression(df.AT, df.PE)
```

```python
print("Here beta2 of X^2 has p-value 0.768, which is greater than 0.05. \
Therefore, we fail to reject the null hypothesis of beta2 =0. \
But X^3 has p-value 0.014 less than 0.05. Therefore, we could reject the null hypothesis of beta3 =0. \
And we conclude that there is evidence of nonlinear association between the predictor V and the response PE")

Polyregression(df.V, df.PE)
```

```python
print("All beta has p-value almost 0, which is less than 0.05. \
Therefore, we could rejetct the null hypothesis and conclude that \
there is evidence of nonlinear association between the predictor RH and the response PE. \n")
Polyregression(df.RH, df.PE)
```

```python
print("All beta has p-value almost 0, which is less than 0.05. \
Therefore, we could rejetct the null hypothesis and conclude that \
there is evidence of nonlinear association between the predictor AP and the response PE. \n")
Polyregression(df.AP, df.PE)
```

### (h) Improvement

```python
# https://realpython.com/knn-python
# https://machinelearningmastery.com/train-test-split-for-evaluating-machine-learning-algorithms/
'''
fixing the seed for the pseudo-random number generator used when splitting the dataset.
1. Model with just the predictors.Â 
2. Model with the predictors + quadratic polynomials + Interactions between predictors.
3. Model with the predictors + quadratic polynomials + Interactions between predictors after removing the insignificant terms.Â 
'''
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.3, random_state=12345 )

train_data = pd.concat([X_train, y_train], axis="columns")
test_data = pd.concat([X_test, y_test], axis = "columns")
```

```python
mse_dic = {}
```

```python
# models with only predictors
formula = "PE ~ AT + V + RH + AP "
model = smf.ols(formula = formula, data = train_data)
result = model.fit()
print(result.summary())

y_pred_test = result.predict(X_test)
y_pred_train = result.predict(X_train)
print("\n")
mse_test =mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error for test data:",mse_test)

mse_train =mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error for train data:",mse_train)

mse_dic["all_predictor_only"] = [mse_train, mse_test]
```

```python
# model involving all possible interaction terms and quadratic nonlinearities
formula = "PE ~ AT + V + RH + AP + np.power(AT,2) + np.power(V,2) + np.power(AP,2) + np.power(RH,2) + \
AT*V + AT*RH + AT*AP + V*RH + V*AP + RH*AP"
model = smf.ols(formula = formula, data = train_data)
result = model.fit()
print(result.summary())
y_pred_test = result.predict(X_test)
y_pred_train = result.predict(X_train)

print("\n")
mse_test =mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error for test data:",mse_test)

mse_train =mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error for train data:",mse_train)


mse_dic["all_interaction_quadratic"] = [mse_train, mse_test]
```

#### From the regression result shown above, the p-values of AT, V, AT:RH, AT:AP, V:RH, V:AP are greater than significant level of 0.05. Therefore, we fail reject the null hypothesis and there is not enough evidence to that prove these terms are statistically signifiant. Also, v^2 term has p-value is 0.05, which is tricky, and I think there is not enough evidence to that prove this term is statistically signifiant.

#### Although AT nad V has large p-value, removing the basic predictors will violate the hierachical principle when AT:V is included.
<br>

```python
# remove the terms from the result above
formula = "PE ~ AT+ V+ RH + AP + np.power(AT,2) + np.power(AP,2) + np.power(RH,2) + AT*V+ RH*AP"
model = smf.ols(formula = formula, data = train_data)
result = model.fit()
print(result.summary())
y_pred_test = result.predict(X_test)
y_pred_train = result.predict(X_train)
print("\n")
mse_test =mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error for test data:",mse_test)

mse_train =mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error for train data:",mse_train)

print("\nAll p-vlaue are less than 0.05. All predictors are statisticall significant.")


mse_dic["remove_all_at_a_time"] = [mse_train, mse_test]
```

```python
# backward selection process for the improved model
'''from the second model with formula 
"PE ~ AT + V + RH + AP + np.power(AT,2) + np.power(V,2) + np.power(AP,2) + np.power(RH,2) + \
AT*V + AT*RH + AT*AP + V*RH + V*AP + RH*AP"
V:RH has highest p-value 0.767 , first remove this
'''
formula = "PE ~ AT + V + RH + AP + np.power(AT,2) + np.power(V,2) + np.power(AP,2) + np.power(RH,2) + \
AT*V + AT*RH + AT*AP + V*AP + RH*AP"
model = smf.ols(formula = formula, data = train_data)
result = model.fit()
print(result.summary())
# AT:AP has the highest p-value 0.386 , remove this

formula = "PE ~ AT + V + RH + AP + np.power(AT,2) + np.power(V,2) + np.power(AP,2) + np.power(RH,2) + \
AT*V + AT*RH + V*AP + RH*AP"
model = smf.ols(formula = formula, data = train_data)
result = model.fit()
print(result.summary())
# all p-values less than 0.05


print("\n")
print("Since all p-values are less than 0.05 after first removing V:RH and then AT:AP,\
we achieve the final model having formula = PE ~ AT + V + RH + AP + np.power(AT,2) + np.power(V,2) + np.power(AP,2) + np.power(RH,2) + \
AT*V + AT*RH + V*AP + RH*AP")

y_pred_test = result.predict(X_test)
y_pred_train = result.predict(X_train)
print("\n")
mse_test =mean_squared_error(y_test, y_pred_test)
print("Mean Squared Error for test data:",mse_test)

mse_train =mean_squared_error(y_train, y_pred_train)
print("Mean Squared Error for train data:",mse_train)


mse_dic["backward selection"] = [mse_train, mse_test]
```

#### Comparing all models, the model with only linear predictors has highest MSE for test data of 20.51. Through improving the model, our MSE for test data decreases to around 18. The blackward selection model has test MSE 17.94 close to the model having all possible interaction terms and quadratic nonlinearities (Test MSE = 17.93) . 

### (i) KNN

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html

# raw data 
k_values=[i for i in range(1,101)]
k_reverse = []
train_errors=[]
test_errors=[]
min_k, min_error = float("inf"), float("inf")

for k in k_values:
    neigh = KNeighborsRegressor(n_neighbors=k)
    neigh.fit(X_train, y_train)
    k_reverse.append(1/k)
    
    y_train_pred = neigh.predict(X_train)
    y_test_pred = neigh.predict(X_test)
    
    train_error = mean_squared_error(y_train, y_train_pred)
    test_error = mean_squared_error(y_test, y_test_pred)
    
    train_errors.append(train_error)
    test_errors.append(test_error)
    
    if (test_error) < min_error:
        min_error = test_error
        min_k = k
        min_train_error = train_error
print(f"Raw features KNN - Best K Value = {min_k} with Test Error = {min_error}")


mse_dic["knn raw"] = [min_train_error, min_error]
```

```python
plt.plot(k_reverse, train_errors, color = "r", label = "Train Error")
plt.plot(k_reverse, test_errors, color = "b", label = "Test Error")
plt.xlabel("K Reverse values")
plt.ylabel("MSE")
plt.title("Train and Test Errors vs K Reverse Values with Raw Data")
plt.legend()
plt.show()
```

```python
# normalized data
# https://www.analyticsvidhya.com/blog/2020/04/feature-scaling-machine-learning-normalization-standardization/

norm = MinMaxScaler().fit(X_train)
X_train_norm = norm.transform(X_train)
X_test_norm = norm.transform(X_test)


k_values=[i for i in range(1,101)]
k_reverse = []
train_errors=[]
test_errors=[]
min_k, min_error = float("inf"), float("inf")

for k in k_values:
    neigh = KNeighborsRegressor(n_neighbors=k)
    neigh.fit(X_train_norm, y_train)
    k_reverse.append(1/k)
    
    y_train_pred = neigh.predict(X_train_norm)
    y_test_pred = neigh.predict(X_test_norm)
    
    train_error = mean_squared_error(y_train, y_train_pred)
    test_error = mean_squared_error(y_test, y_test_pred)
    
    train_errors.append(train_error)
    test_errors.append(test_error)
    
    if (test_error) < min_error:
        min_error = test_error
        min_k = k
        min_train_error = train_error
print(f"Normalized features KNN - Best K Value = {min_k} with Test Error = {min_error}")

mse_dic["knn normalized"] = [min_train_error, min_error]
```

```python
plt.plot(k_reverse, train_errors, color = "r", label = "Train Error")
plt.plot(k_reverse, test_errors, color = "b", label = "Test Error")
plt.xlabel("K Reverse values")
plt.ylabel("MSE")
plt.title("Train and Test Errors vs K Reverse Values with Normalized Data")
plt.legend(loc='lower left')
plt.show()
```

## 2. ISLR: 2.4.1

### (a) The sample size n is extremely large, and the number of predictors p is small.

#### A flexible model will perform better than an inflexible method. Because of the large sample size, weâ€™re less likely to overfit even when using a more flexible model. In addition, a more flexible model tends to reduce bias.

### (b) The number of predictors p is extremely large, and the number of observations n is small.

#### An inflexible model will perform better than a flexible model in this case. A flexible model will cause overfitting because of the small sample size. This indicated a bigger inflation in variance and a small reduction in bias.

### (c) The relationship between the predictors and response is highly non-linear.

#### A flexible model will perform better because itâ€™ll be necessary to use a flexible model to find the non-linear effect.

### (d) The variance of the error terms, i.e. $Ïƒ^2$ = Var(Îµ), is extremely high.

#### An inflexible model is better because when the variance of the error terms is extremely high, our data has significant noise. This might lead to overfitting when using flexible model. Therefore, an inflexible model might be better in this case.

### (a) Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0.

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.euclidean_distances.html
obs = [[0,3,0,'Red'],
       [2,0,0,'Red'],
       [0,1,3,'Red'],
       [0,1,2,'Green'],
       [-1,0,1,'Green'],
       [1,1,1,'Red']]
df1 =pd.DataFrame(obs, columns = ['X1','X2','X3','Y'], index=[1,2,3,4,5,6])
df1

X_train = np.array(df1.iloc[:, :-1])
y_train = np.array(df1.iloc[:, -1])
X_test = np.array([[0, 0, 0]])

dist = euclidean_distances(X_train, X_test)

for i, d in enumerate(dist):
    print(f"The Euclidean distance between obs[{i+1}] and the testpoint X1=X2=X3=0 is {np.round(d,3)}.")
```

### (b) What is our prediction with K = 1? Why?

#### Our prediction with K=1 is Green. With K=1, the test point will has the same prediction result to the observation point closest to it. In thise case, the fifth observation classified to Green has the smalles euclidean distance 1.414 with the test point in all observation points.

### (c) What is our prediction with K = 3? Why?

#### Our prediction with K=1 is Red. We find the response of the nearest 3 points from our test point and do majority polling. In this case, we have observation 5, 6, 2 closest to the test point. We get two Red and one Green. Thus, the prediction result with k=3 is Red.

### (d) If the Bayes decision boundary in this problem is highly non-linear, then would we expect the best value for K to be large or small? Why?

#### A smaller value of K will works better when decision boundary is highly non-linear. When K becomes larger, we get a smoother boundary. Thus, a large value would not be flexible enough to model the nonlinear boundary.
